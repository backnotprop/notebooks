```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats
```

```python
a = np.linspace(0, 4, 100)
exp = stats.expon
lambda_ = [0.5, 1, 1.5]
colors = ["#348ABD", "#A60628", "#008000"]
```

```python
for l, c in zip(lambda_, colors):
    plt.plot(a, exp.pdf(a, scale=1./l), lw="3",
            color=c, label="$\lambda = %.1f$" % l)
    plt.fill_between(a, exp.pdf(a, scale=1./l), color=c, alpha=.33)
    
plt.legend()
plt.ylabel("Probability density function at $z$")
plt.xlabel("$z$")
plt.ylim(0,1.2)
plt.title("Probability density function of exp random variable, differing lambdas")
```
